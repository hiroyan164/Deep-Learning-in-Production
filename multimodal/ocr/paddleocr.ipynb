```python
from paddleocr import PaddleOCR,draw_ocr

from PIL import Image
import requests
```

```python
url = "https://raw.githubusercontent.com/fuyu-quant/data-science-wiki/main/images/logo1.png"
image = Image.open(requests.get(url, stream=True).raw).convert("RGB")
image.save("/content/cat.jpg", "JPEG")
image
```

### Paddleocrの実行
以下の言語に対応している
* 中国語:'ch'
* 英語:'en'
* フランス語:'french'
* ドイツ語:'german'
* 韓国語:'korean'
* 日本語:'japan'

```python
ocr = PaddleOCR(use_angle_cls=True, lang='en')

img_path = '/content/samle.jpg'
result = ocr.ocr(img_path, cls=True)
for idx in range(len(result)):
    res = result[idx]
    for line in res:
        print(line)
```
